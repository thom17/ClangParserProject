```python
#일단 data set 로드
import sys
sys.path.append('../')

import snResult.ClangAST as SNResult

snResult: [SNResult.TClangAST] = SNResult.get_all_table('../snResult/dataSet.db')
```

```python
#data set 설정
update_methods_map: dict[str, SNResult.TClangAST] = {}
for astInfo in snResult:
    srcSig: str = astInfo.srcSig
    if srcSig.__contains__("OnUpdate"):
        update_methods_map[astInfo.get_class_name()] = astInfo
for clsName in update_methods_map:
    ast = update_methods_map[clsName]
    print(clsName, ", ", ast.srcSig)

panoFixtureAST = update_methods_map["CActuatorPanoFixture"]
print(panoFixtureAST.sourceCode)    
#CActuatorPanoFixture 
```

```python
tempCode="""
int Update(int x, int y)
{
    if(x < y)
        return Update(y, x);
    else
        return (x - y) * y;
}
"""

#Clang 으로 파싱
import clangParser.clangParser as Parser

print(tempCode)
print("================="*4)

tempTUNIT = Parser.parse_context(tempCode, False)
tempTUNITCursor = tempTUNIT.cursor
print(tempTUNITCursor)
childs = list(tempTUNITCursor.get_children())
print(childs.__len__())
```

```python
#Clang 으로 파싱
import clangParser.clangParser as Parser

#print(panoFixtureAST.sourceCode)
#print("================="*4)

panoFixtureTU = Parser.parse_context(panoFixtureAST.sourceCode, False)
panoClangCursor = panoFixtureTU.cursor
print(panoClangCursor)
childs = list(panoClangCursor.get_children())
print(childs.__len__())

import clang.cindex
from clangParser.Cursor import Cursor
panoFixtureCursor:Cursor = Cursor(panoFixtureTU.cursor)
panoFixtureCursor.visit_print()
childs = list(panoFixtureCursor.node.get_children())
print(childs.__len__())
# panoFixtureCursor.
```

```python
line_map = panoFixtureCursor.get_visit_line_map()
for num in sorted(line_map):
    print(num, end=" ")
    for c in line_map[num]:
        print(f"<{c.kind} {c.spelling} {c.get_range()}>", end=" ")
    print()
```

```python
stmt_map = panoFixtureCursor.get_visit_stmt_map()

for type_name in stmt_map:
    cursor_list = stmt_map[type_name]
    print(f"{type_name} : {cursor_list.__len__()} size")
```

```python
expr_list = stmt_map["MEMBER_REF_EXPR"]

print(type(expr_list))
for c in expr_list:
    print("="*10)
    print(f"{c.get_range_code()} @{c.location.line}")
    print("="*10)
```

```python
expr_list = stmt_map["CALL_EXPR"]

print(type(expr_list))
for c in expr_list:
    print("="*10)
    print(f"{c.get_range_code()} @{c.location.line}")
    print("="*10)
```
