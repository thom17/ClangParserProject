```python
"""
0 : ref, def 둘중 하나 혹은 둘다 없음
1 : 둘다 있으나 서로 다름 
2 : 둘다 있으며 동일함
"""
print(len(nodes))
same_nodes = []
def get_def_ref(node:clangCursor):
    result_map: [clangCursor, (clangCursor, clangCursor)] = {}
    queue = [node]

    while queue:
        node = queue.pop(0)
        if node in result_map:
            same_nodes.append(node)
            continue
        result_map[node] = (node.get_definition(), node.referenced)

        queue[:0] = node.get_children()
    return result_map

from collections import defaultdict

all_lv_map =defaultdict(list)
def print_result(result_map):
    lv_map = defaultdict(list)
    for node in result_map:
        def_node, ref_node = result_map[node]
        if def_node and ref_node:
            if def_node == ref_node:
                lv_map[2].append(node)
                r_str="lv2"
            else:
                lv_map[1].append(node)
                r_str="lv1"
        else:
            lv_map[0].append(node)
            r_str = "lv0"
                
                
        
        # print(f"{node.kind.name} {node.spelling} {r_str}")

    print(f"0:{len(lv_map[0])} / 1:{len(lv_map[1])} / 2:{len(lv_map[2])}")
    all_lv_map[0][:0] = lv_map[0]
    all_lv_map[1][:0] = lv_map[1]
    all_lv_map[2][:0] = lv_map[2]
for node in nodes:
    result = get_def_ref(node)
    print(CursorOMS(node).get_src_name())
    print_result(result)
    print()

print(f"0:{len(all_lv_map[0])} / 1:{len(all_lv_map[1])} / 2:{len(all_lv_map[2])}")

all_lv_set_map={}
all_lv_set_map[0] = set(all_lv_map[0])
all_lv_set_map[1] = set(all_lv_map[1])
all_lv_set_map[2] = set(all_lv_map[2])

print(f"to set 0:{len(all_lv_set_map[0])} / 1:{len(all_lv_set_map[1])} / 2:{len(all_lv_set_map[2])}")
```

```python
#한쪽만 있거나 둘다 없는 경우 분석.
def_lv0 = set()
ref_lv0 = set()
both = set()

def_lv0_kind = set()
ref_lv0_kind = set()
both_kind = set()


for node in all_lv_map[0]:
    def_node = node.get_definition()
    ref_node = node.referenced

    bAdd = False
    if def_node:
        def_lv0.add(node)
        def_lv0_kind.add(node.kind.name)
        bAdd = True

    bAdd = False
    if ref_node:
        ref_lv0.add(node)
        ref_lv0_kind.add(node.kind.name)
        assert not bAdd, "????" 
        bAdd = True

    if not bAdd:
        both_kind.add(node.kind.name)
        both.add(node)
        
print(f"def vs ref = {len(def_lv0)} vs {len(ref_lv0)} both {len(both)}")
print(f"kind num  = {len(def_lv0_kind)} vs {len(ref_lv0_kind)} both {len(both_kind)}")

for kind in ref_lv0_kind:
    print(kind)
```
