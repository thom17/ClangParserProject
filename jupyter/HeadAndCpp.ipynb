```python
import os
import xml.etree.ElementTree as ET
from clang.cindex import Index, TranslationUnit, TranslationUnitLoadError

def resolve_macros(path, solution_dir):
    path = path.replace('$(SolutionDir)', solution_dir)
    path = path.replace('%(AdditionalIncludeDirectories)', '')
    path = path.replace('%(PreprocessorDefinitions)', '')
    return path

def resolve_macros_in_list(paths, solution_dir):
    return [resolve_macros(path, solution_dir) for path in paths if path]

def extract_compile_options(vcxproj_path, solution_dir):
    tree = ET.parse(vcxproj_path)
    root = tree.getroot()

    include_dirs = []
    defines = []
    additional_options = []

    namespace = {'ns': 'http://schemas.microsoft.com/developer/msbuild/2003'}
    
    for item in root.findall('.//ns:ItemDefinitionGroup/ns:ClCompile', namespace):
        for include in item.findall('ns:AdditionalIncludeDirectories', namespace):
            include_dirs.extend(include.text.split(';'))
        for define in item.findall('ns:PreprocessorDefinitions', namespace):
            defines.extend(define.text.split(';'))
        for option in item.findall('ns:AdditionalOptions', namespace):
            additional_options.extend(option.text.split())

    resolved_include_dirs = resolve_macros_in_list(include_dirs, solution_dir)
    resolved_defines = resolve_macros_in_list(defines, solution_dir)
    
    # Remove any empty strings from the defines list
    resolved_defines = [define for define in resolved_defines if define]

    return resolved_include_dirs, resolved_defines, additional_options

def print_compile_options(include_dirs, defines, additional_options):
    print("Include Directories:")
    for include_dir in include_dirs:
        print(f"  {include_dir}")

    print("\nPreprocessor Definitions:")
    for define in defines:
        print(f"  {define}")

    print("\nAdditional Options:")
    for option in additional_options:
        print(f"  {option}")

def parse_source_file_with_clang(source_file, include_dirs, defines, additional_options):
    args = []
    for include_dir in include_dirs:
        args.append(f'-I{include_dir}')
    for define in defines:
        args.append(f'-D{define}')
    args.extend(additional_options)

    # Debug: Print the arguments passed to Clang
    print("\nClang Arguments:")
    for arg in args:
        print(arg)

    index = Index.create()
    try:
        tu = index.parse(source_file, args=args)
    except TranslationUnitLoadError as e:
        print(f"Error parsing translation unit: {e}")
        return None
    
    return tu

def print_diagnostics(tu):
    if tu is None:
        print("Translation unit could not be loaded.")
        return
    
    for diagnostic in tu.diagnostics:
        print(f"{diagnostic.severity}: {diagnostic.spelling} (line {diagnostic.location.line}, column {diagnostic.location.column})")

def main(solution_dir, source_file, vcxproj_path):
    include_dirs, defines, additional_options = extract_compile_options(vcxproj_path, solution_dir)
    
    # Print the compile options
    print_compile_options(include_dirs, defines, additional_options)
    
    tu = parse_source_file_with_clang(source_file, include_dirs, defines, additional_options)
    print_diagnostics(tu)

    return tu


solution_dir = "D:\\dev\\EcoCad\\trunk\\SimpleTask\\"
source_file = 'D:\\dev\\EcoCad\\trunk\\SimpleTask\\mod_SCCrownDesign\\CommandCrownDesignContact.cpp'  # 파싱할 파일의 경로
vcxproj_path = 'D:\\dev\\EcoCad\\trunk\\SimpleTask\\mod_SCCrownDesign\\mod_SCCrownDesign.vcxproj'  # 해당 파일의 프로젝트 파일 경로

tu=main(solution_dir, source_file, vcxproj_path)


```

```python
def get_call_definition_map(node)->dict[clangCursor, clangCursor]:
    """
    DFS로 target_stmt의 defintion node를 구한다.
    :param target_stmt:
    :return dec_dict: [call_node: clangCursor, def_node: clangCursor]
    """
    dec_dict: [clangCursor, clangCursor] = {}
    queue = [node]

    while queue:
        node = queue.pop(0)
        def_node = node.get_definition()
        if def_node:
            dec_dict[node] = def_node
        queue[:0] = node.get_children()
    return dec_dict

def get_visit_line_map(root_node) -> dict[int, list]:
        line_map = {}
        queue = [root_node]

        while queue:
            node = queue.pop(0)
            line = node.location.line
            if line not in line_map:
                line_map[line] = []
            line_map[line].append(node)
            queue.extend(node.get_children())
        
        return line_map
```
