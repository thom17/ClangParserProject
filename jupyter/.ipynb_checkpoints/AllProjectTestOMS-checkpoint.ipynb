```python
def get_call_definition_map(node)->dict[clangCursor, clangCursor]:
    """
    DFS로 target_stmt의 defintion node를 구한다.
    :param target_stmt:
    :return dec_dict: [call_node: clangCursor, def_node: clangCursor]
    """
    dec_dict: [clangCursor, clangCursor] = {}
    queue = [node]

    while queue:
        node = queue.pop(0)
        def_node = node.get_definition()
        if def_node:
            dec_dict[node] = def_node
        queue[:0] = node.get_children()
    return dec_dict

def get_visit_line_map(root_node) -> dict[int, list]:
        line_map = {}
        queue = [root_node]

        while queue:
            node = queue.pop(0)
            line = node.location.line
            if line not in line_map:
                line_map[line] = []
            line_map[line].append(node)
            queue.extend(node.get_children())
        
        return line_map
```

```python
from clang.cindex import CursorKind
from collections import defaultdict
kind_map = defaultdict(list)


for k in CursorKind.get_all_kinds():
    isin = False
    count = 0 
    if k.is_attribute():
        kind_map["attribute"].append(k.name) 
        isin = True
        count+=1
    if k.is_declaration():
        kind_map["dec"].append(k.name)
        isin = True
        count+=1
    if k.is_expression():
        kind_map["exp"].append(k.name)
        isin = True
        count+=1
    if k.is_reference():
        kind_map["ref"].append(k.name)
        isin = True
        count+=1
    if k.is_statement():
        kind_map["stmt"].append(k.name)
        isin = True
        count+=1
    if not isin:
        kind_map["else"].append(k.name)
    if 1< count:
        kind_map["multi"].append(k.name) 

print(kind_map["exp"])

# for key in kind_map:
#     print(key)
#     print(kind_map[key])
#     print()
    
```
