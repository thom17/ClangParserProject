```python
from clangParser.CUnit import CUnit
from clangParser.Cursor import Cursor
import clangParser.clangParser as clangParser
from clang.cindex import Cursor as clangCursor

target_project_path = r'D:\dev\AutoPlanning\trunk\AP_trunk_pure'
filepath_list = list(clangParser.find_cpp_files(target_project_path))
print('all file : ', len(filepath_list))

all_filepath_list = filepath_list

def is_skip_file(path:str):
    skip_keyword = ['.h', 'dlls', 'meshlibs', 'Test', 'external_libs']
    for key in skip_keyword:
        if key in path:
            return True
    return False

filepath_list = []
for path in all_filepath_list:
    if is_skip_file(path):
        continue
    else:
        filepath_list.append(path)
        
print('filter file : ', len(filepath_list))
```

```python
for path in filepath_list:
    print(path)
```

```python
import time
st = time.time()
my_unit_list = []
for idx, path in enumerate(filepath_list):
    print(f'\r{idx}/{len(filepath_list)} {path}\033[K', end="")
    my_unit_list.append(CUnit.parse(path))
    
ed = time.time()
print(ed-st)
```

```python
from typing import Dict, List
#메서드 하나 단위로 태스트 
def get_context_node(curosr: Cursor, context = 'DeleteFile') -> Dict[int, Cursor]:
    st = time.time()
    line_codes = curosr.get_range_code().splitlines()
    target_line_list = []
    for idx, line in enumerate(line_codes):
        if context in line:
            target_line_list.append(curosr.location.line + idx)
            
    line_map = curosr.get_visit_line_map()
    node_map = {}
    for line in target_line_list:
        for node in line_map[line]:
            if node.get_line_size() == 1 and context in node.get_range_code():
                node_map[line] = node
                break
    ed = time.time()
    return node_map
    
def get_intab(node:Cursor):
    line_code = node.get_range_line_code()
    code_part = node.get_range_code()
    print('line:', line_code)
    print('part:', code_part)
    start_idx = line_code.index(code_part)
    return line_code[:start_idx]

def get_replace_code(method_cursor: Cursor, target_nodes:Dict[int, Cursor], context = 'DeleteFile'):
    replace_code = method_cursor.get_range_code()
    method_code = replace_code.splitlines()
    replaced_line = set()

    for line, node in target_nodes.items():
        index_num = node.location.line- method_cursor.location.line 
        org_line_code = method_code[index_num]

        parm=get_call_exp_parm(node, context)
        # print('org:',org_line_code)
        # print('range:', node.get_range_code())

        intab = get_intab(node)
        replace_exp = '//auto-gen del file\n{' \
          + intab +'LOGN_(System)<<"delete File path: "<<' + parm + ';\n' \
          + org_line_code + '\n}'

        method_code[index_num] = replace_exp

    replace_code = ''
    for line in method_code:
        replace_code += line + '\n'
    return replace_code



def get_call_exp_parm(node: Cursor, context: str):
    que = [node]
    node = que.pop(0)
    while node.kind != 'CALL_EXPR' or not context in node.spelling:
        print(node)
        que = que + [Cursor(n) for n in node.node.get_children()]
        if que:
            node = que.pop(0)
        else:
            return None

    code = node.get_range_code()
    first_open = code.index('(')
    return code[first_open:]

# target = Cursor(clang_src_map['CFixtureData.DeleteFixture(bool)'][0])
# line_node = get_context_node(target)
# replace_code=get_replace_code(target, line_node)

# print('time ', ed-st)
# print(replace_code)
```

```python
from collections import defaultdict
st = time.time()
target_nodes = []
delete_method_dict = defaultdict(list)
for idx, cunit in enumerate(my_unit_list):
    print(f'\r{idx}/{len(my_unit_list)} {cunit.file_path}\t\t\t\t', end="")
    for node in cunit.this_file_nodes:        
        cursor = Cursor(node)
        if '::DeleteFile' in cursor.get_range_code():
            delete_method_dict[cunit].append(cursor)
            target_nodes.append(cursor)
ed = time.time()
print(ed-st)
```

```python

def autogen_target_cpp(target_node_map: Dict[CUnit, List[Cursor]], context = 'DeleteFile'):
    for cunit, target_methods in target_node_map.items():
        result_dict = {}
        if target_methods:
            cpp_file_path = cunit.file_path
            cpp_code, _ = target_methods[0].read_file()

            #각 메서드에 대하여 수정할 노드 찾기
            for method_node in target_methods:
                #[라인, 노드]
                line_node_map = get_context_node(curosr=method_node, context=context)
                replace_method_code = get_replace_code(method_cursor=method_node, context=context, target_nodes=line_node_map)

                org_method_code = method_node.get_range_code()
                cpp_code = cpp_code.replace(org_method_code, replace_method_code)

            #헤더 추가
            if not 'plogger.h"' in cpp_code:
                cpp_code:str
                last_include_idx = cpp_code.rfind('#include "')
                cpp_code = cpp_code[:last_include_idx] + '#include "../BaseTools/plogger.h" //auto-gen\n"' + cpp_code[last_include_idx:] 
            result_dict[cpp_file_path] = cpp_code      
        return result_dict

result_dict = autogen_target_cpp(delete_method_dict)
```

```python
for filepath, code in result_dict.items():
    with open(filepath, 'w') as file:
        file.write(code)
print('done')
```
