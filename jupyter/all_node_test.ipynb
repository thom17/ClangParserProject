```python
from clang.cindex import Token, TokenKind
from clangParser.Cursor import Cursor as MyCursor
def get_src_name(node: clangCursor):
    kind:CursorKind = node.kind
    def_node = node.get_definition()
    if kind.is_declaration():
        name = __get_def_name(node)
    else:
        name = __get_exp_name(node)


    if __is_root_node(node):
        return name    
    else:
        return get_src_name(node.semantic_parent) + "." +name

def __is_root_node(node:clangCursor):
    parent = node.semantic_parent
    if parent is None or parent.kind == CursorKind.TRANSLATION_UNIT:
        return True
    else:
        return False
        
def __get_exp_name(node: clangCursor):
    #연산과 관련된 식
    return MyCursor(node).get_range_code() + ":" + str(node.location.line)

def __get_def_name(node: clangCursor):
    #메서드의 경우
    if node.kind.name in ["CXX_METHOD", "CONSTRUCTOR" ,"DESTRUCTOR", "FUNCTION_DECL"]:
        return node.spelling + __get_argument_sig(node)
    #단순한 타입의 경우 (변수)
    else:
        return node.spelling

def __get_argument_sig(node: clangCursor):
    result_sig = ""
    for dec_node in node.get_arguments():
        # dec_oms = CursorOMS(dec_node)
        # dec_code = dec_oms.get_range_code()

        parm_sig =__get_param_sig_by_token(dec_node)
        result_sig+= ", " + parm_sig

    if result_sig:
        return "(" + result_sig[2:]+")"
    else:
        return "()"



def __get_param_sig_by_token(node: clangCursor):
    """
    token을 사용하여 PARM_DECL 하나의 시그니쳐를 구함.
    blabla type id (= blabla )에서 ()를 날림.
    """
    assert node.kind == CursorKind.PARM_DECL, f"파라미터 이름을 구하는 용도. {node.kind} 입력됨"
    result = ""
    last_token_type = TokenKind.PUNCTUATION
    for token in node.get_tokens():
        token: Token 
        if token.kind != TokenKind.PUNCTUATION:
            if token.spelling == node.spelling: #종료 지점 찾음. 타입 정보만 필요함
                break 
            elif last_token_type == TokenKind.PUNCTUATION : #:: < & * 와 같은 기호들은 붙여서 출력
                result += token.spelling
            else: #그 외는 한칸 공백
                result += " "+token.spelling
        else: #식별자가 아니라면 그냥 더한다.
            result += token.spelling
            
        last_token_type = token.kind #공백 여부를 위해 아무튼 마지막 토큰 타입 저장
    return result


axial_init=untis[0].this_file_nodes[2]
# print(CursorOMS(axial_init).get_range_code())

from collections import defaultdict
src_map = {}

for node in nodes:
    src_name = get_src_name(node)
    if not src_name in src_map:
        src_map[src_name] = set()
    src_map[src_name].add(node)
print(len(nodes), end="->")
print(len(src_map))

for key in src_map:
    print(f"{key} {len(src_map[key])}")

# for child in axial_init.get_children():
#     print(child.kind)
#     print("\t",child.spelling)

cls_def = axial_init.semantic_parent 
cls_def==cls_def.get_definition()
cls_parent = cls_def.semantic_parent
print(cls_parent.kind, "\t", cls_parent.spelling)


# print(cls_def.spelling,"\t",cls_def.kind)
# for child in cls_def.get_children():
#     print(child.kind)
#     print("\t",child.spelling)

# print(last_cursor.kind)
# print(last_cursor.location)
# args = last_cursor.get_arguments()
# for compound in last_cursor.get_children():
#     print(f"{compound.spelling} {compound.kind.name}")
#     print(CursorOMS(compound).get_range_code())
#     for child in compound.get_children():
#         print(f"{child.spelling} {child.kind.name}")
```

```python
from clang.cindex import CursorKind
def get_methods(cursor_list: [clangCursor]) -> [str, [clangCursor]]:
    result_map = defaultdict(list)
    for node in cursor_list:
        kind:CursorKind = node.kind
        # kind.is
        node: clangCursor
        added = 0
        if node.is_abstract_record():
            result_map["is_abstract_record"].append(node)
            added+=1
        if node.is_anonymous():
            result_map["is_anonymous"].append(node)
            added+=1
        if node.is_bitfield():
            result_map["is_bitfield"].append(node)
            added+=1
        if node.is_const_method():
            result_map["is_const_method"].append(node)
            added+=1
        if node.is_converting_constructor():
            result_map["is_converting_constructor"].append(node)
            added+=1
        if node.is_copy_assignment_operator_method():
            result_map["is_copy_assignment_operator_method"].append(node)
            added+=1
        if node.is_copy_constructor():
            result_map["is_copy_constructor"].append(node)
            added+=1
        if node.is_default_constructor():
            result_map["is_default_constructor"].append(node)
            added+=1
        if node.is_default_method():
            result_map["is_default_method"].append(node)
            added+=1
        if node.is_definition():
            result_map["is_definition"].append(node)
            added+=1
        if node.is_deleted_method():
            result_map["is_deleted_method"].append(node)
            added+=1
        if node.is_explicit_method():
            result_map["is_explicit_method"].append(node)
            added+=1
        if node.is_move_assignment_operator_method():
            result_map["is_move_assignment_operator_method"].append(node)
            added+=1
        if node.is_move_constructor():
            result_map["is_move_constructor"].append(node)
            added+=1
        if node.is_mutable_field():
            result_map["is_mutable_field"].append(node)
            added+=1
        if node.is_pure_virtual_method():
            result_map["is_pure_virtual_method"].append(node)
            added+=1
        if node.is_scoped_enum():
            result_map["is_scoped_enum"].append(node)
            added+=1
        if node.is_static_method():
            result_map["is_static_method"].append(node)
            added+=1
        if node.is_virtual_method():
            result_map["is_virtual_method"].append(node)
            added+=1

    return result_map
```

```python
#모든 노드 타입에 대한 def node 구하는 함수태스트
def_lv_map = defaultdict(list)
def get_def_node(node:clangCursor):
    node_def = node.get_definition()
    node_ref = node.referenced
    if node_def:
        def_lv_map[3].append(node)
        return node_def
    elif node_ref and node_ref.get_definition():
        def_lv_map[2].append(node)
        return node_ref.get_definition()
    else:
        if node_ref:
            def_lv_map[1].append(node)
        else:
            def_lv_map[0].append(node)
        
        return node_ref

def_node_map={}
for node in all_visit_nodes:
    def_node_map[node] = get_def_node(node)

lv3_count = len(def_lv_map[3])
lv2_count = len(def_lv_map[2])
lv1_count = len(def_lv_map[1])
lv0_count = len(def_lv_map[0])

all_count = len(all_visit_nodes)

print(f"lv3 {lv3_count}({lv3_count*100/all_count}%)")
print(f"lv2 {lv2_count}({lv2_count*100/all_count}%)")
print(f"lv1 {lv1_count}({lv1_count*100/all_count}%)")
print(f"lv0 {lv0_count}({lv0_count*100/all_count}%)")

for kind_name in set([node.kind.name for node in def_lv_map[0]]):
    print(kind_name)
```
